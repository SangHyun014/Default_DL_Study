```python
import numpy as np
import matplotlib.pyplot as plt

import cv2

import glob
import random
```

```python
img = cv2.imread('./s1.jpg' , cv2.IMREAD_COLOR)
```

```python
print(img.shape)
img_1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_1)
```

```python
def fill(img, height, width):
    img = cv2.resize(img, (height, width), cv2.INTER_AREA)
    return img
```

```python
img_r = cv2.resize(img_1 , ( 1440 , 1440 ) , cv2.INTER_AREA)
```

```python
plt.imshow(img_r)
```

```python
def rotation(img, angle):
    angle = int(random.uniform(-angle, angle))
    print(random.uniform(-angle, angle))
    height, width = img.shape[:2]
    M= cv2.getRotationMatrix2D((int(width/2), int(height/2)), angle, 1)
    img = cv2.warpAffine(img, M, (width,height))
    img = fill(img, height, width)
    return img
```

```python
img_ro = rotation(img_r , 30)
```

```python
plt.imshow(img_ro)
```
