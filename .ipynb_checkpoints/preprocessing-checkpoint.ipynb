```python
print(img.shape) #우리가 다루려는 데이터 : 세로 1440 pixel, 가로 1080 pixel
```

```python
def fill(img, height, width):
    img = cv2.resize(img, (height, width), cv2.INTER_AREA)
    return img
```

```python
img_resize = fill(img, 1440, 1440)
```

```python
plt.imshow(img_resize)
```

```python
def rotation(img, angle):
    angle = int(random.uniform(-angle, angle))
    height, width = img.shape[:2] #1440 1440
    M = cv2.getRotationMatrix2D((int(width/2), int(height/2)), angle, 1) #이동을 나타내는 변환행렬 생성
    img = cv2.warpAffine(img, M, (width, height)) #변환 행렬을 Affine 변환 하는 매소드
    img = fill(img, height, width)
    return img
```

```python
img_rotation = rotation(img_resize, 30)
```

```python
plt.imshow(img_rotation)
```
